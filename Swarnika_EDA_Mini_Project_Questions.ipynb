```python
##E. Calculate IQR(the difference between 75% and 25% quartile) for Hourly_Income and Hourly Expense
#Hourly_Income
Q3 = np.quantile(df['Hourly_Income'], 0.75)
Q1 = np.quantile(df['Hourly_Income'], 0.25)
IQR = Q3 - Q1
print(IQR)
```

```python
#Hourly_Expense
Q3 = np.quantile(df['Hourly_Expense'], 0.75)
Q1 = np.quantile(df['Hourly_Expense'], 0.25)
IQR = Q3 - Q1
print(IQR)
```

```python
df = pd.read_csv("Pri-Load.csv")
categories = {
    1: 'Debt Consolidation', 
    2: 'Home Improvement', 
    3: 'Business', 
    6: 'Auto', 
    7: 'Other'
}

def func(categories, col_name):
    col = df[col_name] #extract the column names
    col = col.map(categories) #replace the items in the column with the categories values
    col = col.dropna() #drop that NaN values
    print("Sample Output:\n") #Display
    print(col.value_counts().to_string())

func(categories, 'ListingCategory (numeric)')
```

```python
##C. Display the box plot for ProsperRating (Alpha) vs LoanOriginalAmount and hue = Loan Status[Completed,Defaulted]. Write your observations.
plt.figure(figsize = (10,10)) #Set the figure size
sns.boxplot(x = 'ProsperRating (Alpha)', y = 'LoanOriginalAmount',
           hue = 'LoanStatus', data = df[df['LoanStatus'].isin(['Completed','Defaulted'])])
plt.yticks(ticks = np.arange(0,25001,1000)) #Add the ticks
plt.show()
```

```python
##D. Display the catplot for ProsperRating (Alpha) vs ListingCategory (numeric)[Debt Consolidation,Other,Business,Home Improvement,Auto] and hue = Loan Status[Completed,Defaulted]. Write your observations.
df = pd.read_csv("Pri-Load.csv")
df['ListingCategory (numeric)'] = df['ListingCategory (numeric)'].map(categories)
df = df.dropna()
sns.catplot(x = 'ProsperRating (Alpha)', y = 'ListingCategory (numeric)', hue = 'LoanStatus',
           data = df[df['LoanStatus'].isin(['Completed','Defaulted'])], height=6, aspect=11/6)
plt.show()
```
